```python
# imports
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# parameters
TICKERS = ["AAPL", "GOOGL", "MSFT"]
START_DATE = "2024-01-01"
END_DATE = "2024-12-31"
```

```python
# download prices
print(f"Downloading data for: {', '.join(TICKERS)}")
prices = yf.download(TICKERS, start=START_DATE, end=END_DATE, auto_adjust=True)

prices.head()
```

```python



# Extract closing prices into a clean DataFrame
df = prices["Close"]
df.columns.name = None  # Remove the "Ticker" label from columns
print(f"\nDataFrame shape: {df.shape}")
print(df.head())

# --- Normalize to 100 at start (for easy comparison) ---
df_normalized = (df / df.iloc[0]) * 100

# --- Plot ---
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Top chart: raw closing prices
for ticker in TICKERS:
    axes[0].plot(df.index, df[ticker], label=ticker, linewidth=1.5)

axes[0].set_title("Stock Closing Prices", fontsize=14, fontweight="bold")
axes[0].set_ylabel("Price (USD)")
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
axes[0].xaxis.set_major_locator(mdates.MonthLocator(interval=2))

# Bottom chart: normalized to 100 for performance comparison
for ticker in TICKERS:
    axes[1].plot(df_normalized.index, df_normalized[ticker], label=ticker, linewidth=1.5)

axes[1].axhline(y=100, color="gray", linestyle="--", linewidth=0.8, alpha=0.7)
axes[1].set_title("Normalized Performance (Base = 100)", fontsize=14, fontweight="bold")
axes[1].set_ylabel("Indexed Value")
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
axes[1].xaxis.set_major_locator(mdates.MonthLocator(interval=2))

plt.tight_layout()
plt.savefig("stock_prices.png", dpi=150, bbox_inches="tight")
print("\nPlot saved to stock_prices.png")
plt.show()

# --- Summary stats ---
print("\n--- Summary Statistics ---")
print(df.describe().round(2))
```
